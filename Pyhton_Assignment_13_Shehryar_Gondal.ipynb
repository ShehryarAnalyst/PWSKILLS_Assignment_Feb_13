__Q1. Explain why we have to use the Exception class while creating a Custom Exception?__<br>
__Note: Here Exception class refers to the base class for all the exceptions.__

__Ans)__ To inherit from the Exception class when creating a custom exception. This allows the custom exception to inherit all the standard behaviors and properties of the Exception class and take advantage of the standard exception handling mechanisms provided by the language, such as the try-except statement. In addition, a custom exception can define its own behaviors and properties as needed. By inheriting from the Exception class, a custom exception can be caught and handled in the same way as other built-in exceptions.

__Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.__

__Ans)__ The ArithmeticError class is a built-in Python exception class that is raised for various arithmetic errors. It is a subclass of the Exception class and is itself the parent class for a number of more specific arithmetic exceptions.<br>
__Example__: <br>
1. ZeroDivisionError: This exception is raised when attempting to divide a number by zero.
2. OverflowError: This exception is raised when the result of an arithmetic operation exceeds the maximum representable value.

```python
# Example of ZeroDivisionError
try:
    result = 1 / 0
except ZeroDivisionError as e:
    print(f"Error: {type(e).__name__} - {e}")
```

```python
# Example of OverflowError
try:
    result = 10 ** 1000
except OverflowError as e:
    print(f"Error: {type(e).__name__} - {e}")
```

__Q4. Why LookupError class is used? Explain with an example KeyError and IndexError ? __

__Ans)__ The LookupError class is a built-in Python exception class that is raised when an index or key lookup fails in a sequence or mapping.<br>
__Example__ <br>
1. __KeyError__: This exception is raised when attempting to access a non-existent key in a dictionary.
2. __IndexError__: This exception is raised when attempting to access a non-existent index in a sequence, such as a list or tuple.

```python
# Example of KeyError
my_dict = {"Shehryar": 1, "Gondal": 2, "Ali": 3}
try:
    value = my_dict["d"]
except KeyError as e:
    print(f"Error: {type(e).__name__} - {e}")
```

```python
# Example of IndexError
my_list = [33, 32, 23]
try:
    value = my_list[3]
except IndexError as e:
    print(f"Error: {type(e).__name__} - {e}")
```

```python
#always try to log your error
import logging
logging.basicConfig(filename = "error.log" , level = logging.ERROR)
try :
    10/0
except ZeroDivisionError as e :
    logging.error("i am trying to handle a zerodivision error {} ".format(e) )
```
